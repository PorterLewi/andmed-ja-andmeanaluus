```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file, skipping rows with invalid string formats
df = pd.read_csv('/kaggle/input/imdb-dataset-of-top-1000-movies-and-tv-shows/imdb_top_1000.csv', error_bad_lines=False)

# Convert the Release Date column to datetime
df['Released_Year'] = pd.to_datetime(df['Released_Year'], errors='coerce')

# Drop rows with missing Release Date values
df.dropna(subset=['Released_Year'], inplace=True)

# Extract the year from the Release Date column
df['Released_Year'] = df['Released_Year'].dt.year

# Count the number of movies released per year
movies_per_year = df['Released_Year'].value_counts().sort_index()

plt.figure(figsize=(12, 6))
plt.bar(movies_per_year.index, movies_per_year.values)
plt.xlabel('Year')
plt.ylabel('Number of Movies Released')
plt.title('Number of Movies Released per Year')
plt.xticks(rotation=45)
plt.show()

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/kaggle/input/imdb-dataset-of-top-1000-movies-and-tv-shows/imdb_top_1000.csv')

# Compute the correlation matrix
correlation_matrix = df.corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the IMDb dataset
df = pd.read_csv('/kaggle/input/imdb-dataset-of-top-1000-movies-and-tv-shows/imdb_top_1000.csv')

# Count the occurrences of each certificate
certificate_counts = df['Certificate'].value_counts()

# Filter the certificate counts to exclude values with less than 1%
certificate_counts_filtered = certificate_counts[certificate_counts >= certificate_counts.sum() * 0.01]

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(certificate_counts_filtered, labels=certificate_counts_filtered.index, autopct='%1.1f%%')
plt.title('Filmide vanusepiirangud')

# Adjust the layout
plt.tight_layout()

# Display the pie chart
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the IMDb dataset
df = pd.read_csv('/kaggle/input/imdb-dataset-of-top-1000-movies-and-tv-shows/imdb_top_1000.csv')

# Count the occurrences of each certificate
certificate_counts = df['Certificate'].value_counts()

# Filter the certificate counts to include values under 1%
certificate_counts_filtered = certificate_counts[certificate_counts < certificate_counts.sum() * 0.01]

# Create a pie chart for values under 1%
plt.figure(figsize=(8, 8))
plt.pie(certificate_counts_filtered, labels=certificate_counts_filtered.index, autopct='%1.1f%%')
plt.title('Filmide vanusepiirangud (Alla 1%)')

# Adjust the layout
plt.tight_layout()

# Display the pie chart
plt.show()
```
